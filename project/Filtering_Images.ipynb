```python
lst = ['無中風', '缺血', '出血']
a_or_b = {'a': [2, [180, 130]], 'b': [1, [100, 50]], 'X': [1, [100, 50]], 'Y': [1, [180, 110]], 'Z': [1, [140, 60]]}
avail = ['a', 'b', 'X', 'Y', 'Ｚ']
for ele in lst:
  files = glob.glob(f"/content/drive/MyDrive/Colab Notebooks/{ele}/*.jpg")

  for file in files:
    ############################### REMOVING OBJECTS #################################
    print(f"Processing {file}...")
    img = Image.open(file)
    img_rgb = np.array(img)

    type_ = file.split('/')[-1][0] # 'a' or 'b'
    print(file)
    img_filtered = img_rgb.copy()

    mask = img_filtered[:, :, 0] > a_or_b[type_][1][0]
    img_filtered[mask, :] = 255

    mask = img_filtered[:, :, a_or_b[type_][0]] > a_or_b[type_][1][1]
    img_filtered[mask, :] = 255

    if type_ == 'Y':
      mask = img_filtered[:, :, 2] > 130
      img_filtered[mask, :] = 255
    # fig = plt.figure(figsize=(20, 10))
    # plt.imshow(img_filtered)
    # plt.axis("off")

    img = Image.fromarray(img_filtered)


    ################################ RESIZE ##############################
    target_w = 1024
    target_h = 640

    w, h = img.size
    img_resized = img.resize((target_w, int(h * target_w / w)), resample=Image.BILINEAR)
    img_padded = 255 * np.ones((target_h, target_w), dtype=np.uint8)
    img_padded[0:img_resized.size[1], 0:img_resized.size[0]] = np.array(img_resized)[:, :, 0]

    img = Image.fromarray(img_padded)

    # img.save('/content/drive/MyDrive/Colab Notebooks/Processed/' + ele + '/' + file.split('/')[-1])
    # print("Saved to... " + '/content/drive/MyDrive/Colab Notebooks/Processed/' + ele + '/' + file.split('/')[-1])
```
